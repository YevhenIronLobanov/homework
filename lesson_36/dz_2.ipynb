```python
import pandas as pd
import numpy as np
from sklearn.model_selection import StratifiedKFold, KFold, cross_val_score
from sklearn import linear_model
```

```python
petrol = pd.read_csv('petrol_consumption.csv')
petrol.head(10)
```

```python
petrol.columns = ['Податок на бензин', 'Середній заробіток', 'Асфальтоване шосе', 'Водійські права', 'Споживання бензину']
petrol.head(10)
```

```python
# Визначення цільової змінної
y = petrol['Споживання бензину']
```

```python
# Видалення цільової змінної 
petrol.drop(['Споживання бензину'], axis=1, inplace=True)
```

```python
# Для уникнення наявності текстових даних створюємо список, який містить імена столбців, що мають числовий тип даних
numeric_cols =[cname for cname in petrol.columns if petrol[cname].dtype in ['int64', 'float64']]
X =petrol[numeric_cols].copy()
X.head(10)
```

```python
X.shape
```

```python
# Cтворюємо об'єкт класа KFold
kf = KFold (n_splits=5, random_state=42, shuffle=True)
cnt = 1
for train_index, test_index in kf.split(X, y):
    print(f'Fold:{cnt}, Train set: {len(train_index)}, Test set: {len(test_index)}')
    cnt +=1
```

```python
# Розраховуємо корінь від'ємної середньоквадратичної похибки 
def rmse(score):
    rmse = np.sqrt(-score)
    print(f'rmse= {"{:.2f}".format(rmse)}')
```

```python
score = cross_val_score(linear_model.LinearRegression(), X, y, cv= kf, scoring="neg_mean_squared_error")
score = cross_val_score(linear_model.LinearRegression(), X, y, cv= kf, scoring="neg_mean_squared_error")
print(f"Від'ємна cередньоквадратична похибка по кожному фолду : {score}")
rmse(score.mean())
```
