```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.metrics import mean_squared_error
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures
from sklearn.pipeline import Pipeline
```

```python
pos_sail = pd.read_csv('Position_Salaries.csv')
pos_sail
```

```python
plt.figure(figsize=(10,5))
plt.title('Рівень заробітної плати в залежності від рівня посади')
plt.xlabel('Рівень посади')
plt.ylabel('Заробітна плата')
plt.xticks(range(1, int(pos_sail['Level'].max()) + 1))
plt.scatter(pos_sail['Level'], pos_sail['Salary'])

           
```

```python
X = np.array([pos_sail['Level']])
y = np.array([pos_sail['Salary']])
X = X.reshape(-1,1)
y = y.reshape(-1,1)
model_lr = LinearRegression()
model_lr.fit(X, y)
```

```python
y_pred= model_lr.predict(X)
plt.title('Рівень заробітної плати в залежності від рівня посади')
plt.xlabel('Рівень посади')
plt.ylabel('Заробітна плата')
plt.xticks(range(1, int(pos_sail['Level'].max()) + 1))
plt.scatter(pos_sail['Level'], pos_sail['Salary'])
plt.plot(X, y_pred,color = 'g')
```

```python
print('RMSE for Linear Regression=>',np.sqrt(mean_squared_error(y,y_pred)))
```

```python
model_Pol =[('polinomial', PolynomialFeatures(degree=3)), ('model', LinearRegression())]
pipe = Pipeline(model_Pol)
pipe.fit(X, y)
```

```python
poly_pred=pipe.predict(X)
sorted_zip = sorted(zip(X,poly_pred))
X_poly, poly_pred = zip(*sorted_zip)
```

```python
plt.figure(figsize=(10,5))
plt.title('Поліноміальна регресія заробітної плати в залежності від рівня посади')
plt.xlabel('Рівень посади')
plt.ylabel('Заробітна плата')
plt.xticks(range(1, int(pos_sail['Level'].max()) + 1))
plt.scatter(pos_sail['Level'], pos_sail['Salary'])
plt.plot(X, y_pred,color = 'g', label = 'Linear Regression')
plt.plot(X_poly, poly_pred,color = 'b', label = 'Polynomial Regression')
plt.legend()
```

```python
print('RMSE for Polynomial Regression=>',np.sqrt(mean_squared_error(y,poly_pred)))
```
