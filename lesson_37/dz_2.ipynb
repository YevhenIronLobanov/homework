```python
import numpy as np
import pandas as pd
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
```

```python
data = pd.read_csv('communities.csv')
data
```

```python
# Предобробка даних
data = pd.read_csv('communities.csv', header=None)
data.columns = [f'Feature_{i}' for i in range(1, len(data.columns))] + ['Target']
data.replace('?', np.nan, inplace=True)
column_order = ['Feature_4'] + [col for col in data.columns if col != 'Feature_4']
data = data[column_order]
data.insert(0, 'Name_city', data.pop('Feature_4'))
data.columns = ['Name_city', 'Target'] + [f'Feature_{i}' for i in range(2, len(data.columns))]
data.fillna(0, inplace=True)

data
```

```python
#Визначення X (ознаки) та y (цільова змінна)
X = data.iloc[:, 2:].values  # Преобразуем DataFrame в numpy.array
y = data.iloc[:, 1].values
```

```python
# Розділяємо на тренувальний та  тестовий набори
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=0)
```

```python
# Нормалізація ознак
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

### L1 pегулярізація даних

```python
from sklearn.linear_model import Lasso
```

```python
#Пошук оптимального alpha за допомогою перехресної крос-валідації
lasso_cv = LassoCV(cv=5, max_iter=10000)  
lasso_cv.fit(X_train_scaled, y_train)
optimal_alpha = lasso_cv.alpha_
optimal_alpha
```

```python
lasso = Lasso(alpha=optimal_alpha).fit(X_train_scaled, y_train)
print(f"Lasso Regression-Training set score: {lasso.score(X_train_scaled, y_train):.2f}")
print(f"Lasso Regression-Test set score: {lasso.score(X_test_scaled, y_test):.2f}")
```

### L2 pегулярізація даних

```python
from sklearn.linear_model import Ridge
from sklearn.linear_model import RidgeCV
```

```python
# Генеруємо послідовність альфа в діапазоні
alphas = np.logspace(-6, 6, 13)  
ridge_cv = RidgeCV(alphas=alphas, store_cv_values=True)
ridge_cv.fit(X_train_scaled, y_train)
optimal_alpha_r = ridge_cv.alpha_
optimal_alpha_r
```

```python
ridge = Ridge(alpha=0.7).fit(X_train_scaled, y_train)
print(f"Ridge Regression-Training set score: {ridge.score(X_train_scaled, y_train):.2f}")
print(f"Ridge Regression-Test set score: {ridge.score(X_test_scaled, y_test):.2f}")
```

### Еластична сітка

```python
from sklearn.linear_model import ElasticNet
from sklearn.linear_model import ElasticNetCV
```

```python
# Пошук оптимального alpha за допомогою перехресної крос-валідації для ElasticNet
elastic_net_cv = ElasticNetCV(
    l1_ratio=np.linspace(0.01, 1, 50),  
    cv=5, n_alphas=100, max_iter=10000, random_state=42)
elastic_net_cv.fit(X_train_scaled, y_train)
optimal_alpha = elastic_net_cv.alpha_
optimal_l1_ratio = elastic_net_cv.l1_ratio_
print(f"Optimal alpha: {optimal_alpha:.4f}")
print(f"Optimal L1 ratio: {optimal_l1_ratio:.4f}")
```

```python
elastic_net = ElasticNet(alpha=optimal_alpha, l1_ratio=optimal_l1_ratio).fit(X_train_scaled, y_train)
print(f"Elastic Net-Training set score: {elastic_net.score(X_train_scaled, y_train):.2f}")
print(f"Elastic Net-Test set score: {elastic_net.score(X_test_scaled, y_test):.2f}")
```
