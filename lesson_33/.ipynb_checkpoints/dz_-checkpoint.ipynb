```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
avto = pd.read_excel('avto.xlsx')
avto
```

```python
print(avto.columns)
```

### *Візуалізація залежностей характеристик автомобіля.*

```python
fig = plt.figure (figsize = (9,7), dpi = 90)
```

```python
plt.title ('Залежність споживання палива від вартості авто.')
plt.xlabel('Fuel consumption(l/100 km)')
plt.ylabel('Price (USD)')
plt.plot(avto['Fuel  (l/100 кm)'], avto ['Price (USD)'],linewidth = 2, markersize = 2, color = 'green')
```

```python
plt.title ('Залежність споживання палива від швидкості авто.')
plt.xlabel('Fuel consumption(l/100 km)')
plt.ylabel('Max_speed (km/h)')
plt.scatter(avto['Fuel  (l/100 кm)'], avto ['Max_speed (km/h)'],  marker = 'D', color = 'blue')
```

```python
plt.title ('Залежність вантажопідйомності від швидкості авто.')
plt.xlabel('Carrying_capacity (kg)')
plt.ylabel('Max_speed (km/h)')
plt.scatter(avto['Carrying_capacity (kg)'], avto ['Max_speed (km/h)'],  marker = '*', color = 'red')
```

### *Розподіл автомобілів по групам*

```python
plt.title ('Розподіл авто по групам в залежності від швидкості.')
plt.xlabel('Max_speed (km/h)')
plt.ylabel('Number of cars')
plt.hist(avto['Max_speed (km/h)'],5, width=4, color = 'yellow')
```

### *Коробковий графік відображення швидкості різних марок автомобілів.*

```python
sns.boxplot(x='Max_speed (km/h)', y= 'Car_model', data = avto)
```

### *Парний графік відображення залежностей величин датасету*

```python
sns.pairplot(avto)
```
