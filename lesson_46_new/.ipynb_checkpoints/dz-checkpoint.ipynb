```python
# Функція сезонності
def sesonal(data, s):
    plt.figure(figsize=(19, 8), dpi=80)
    for i, y in enumerate(data.index.year.unique()):
        plt.plot(
            list(range(1, len(data[data.index.year == y]) + 1)),
            data[data.index.year == y][data.columns[0]].values,
            label=y
        )
    plt.title("Сезонність по періодам")
    
    # Розмістити легенду праворуч ззаду графіка
    plt.legend(loc="upper right", bbox_to_anchor=(1.1, 1))
    
    plt.show()
```

```python
# Функція розрахунку метрик
def metrics(real, forecast):
    
    if type(real)==pd.core.frame.DataFrame:
        real=real[real.columns[0]].values
    
    print("Тест на стаціонарність:")
    dftest = adfuller(real-forecast, autolag='AIC')
    print("\tT-статистика = {:.3f}".format(dftest[0]))
    print("\tP-значення = {:.3f}".format(dftest[1]))
    print("Критичне значення :")
    for k, v in dftest[4].items():
        print("\t{}: {} - Дані {} стационарні з ймовірністю {}% відсотків".format(k, v, "не" if v<dftest[0] else "", 100-int(k[:-1])))
    
    forecast=np.array(forecast)
    print('MAD:', round(abs(real-forecast).mean(),4))
    print('MSE:', round(((real-forecast)**2).mean(),4))
    print('MAPE:', round((abs(real-forecast)/real).mean(),4))
    print('MPE:', round(((real-forecast)/real).mean(),4))
    print('Стандартна похибка:', round(((real-forecast)**2).mean()**0.5,4)) 
```

## Виявлення аномалій та їх коригування

#### *Виявлення аномалій на основі встановленого порогу (метод "ThresholdAD" )*

```python
df.describe()
```

```python
# Розрахунок нижнього та верхнього порога для виявлення аномалій за допомогою міжквартильного діапазону 
Q_1 = 328.570000
Q_3 = 375.990000
IQR = Q_3-Q_1
low_tr=Q_1-1.5*IQR
high_tr=Q_3+1.5*IQR
high_tr, low_tr
```

```python
thrashhold_ad = ThresholdAD(high=447.12, low=257.44)
anomalies = thrashhold_ad.detect(df)
```

```python
# Візуалізація виявлених аномалій 
plot(df, anomaly=anomalies, ts_linewidth=3, ts_markersize=3, anomaly_markersize=5, anomaly_color='red', anomaly_tag='marker')
```

На основі отриманих результатів можна зробити висновок, що в даному часовому ряді аномалії відсутні.

#### *Виявлення аномалій на основі врахування відносного відхилення об'єкта від його сусідів в наборі даних*

```python
outluer = OutlierDetector(LocalOutlierFactor(contamination=0.05))
anomalies = outluer.fit_detect(df)
plot(df,anomaly=anomalies, ts_linewidth=2, anomaly_color='red', anomaly_alpha=0.3, curve_group='all')
```

# Модель Хольта-Вінтерса

Враховуючи результати декомпозиції при побудові моделі використаємо мультиплікативний тренд та сезонність.

```python
fit_1 = ExponentialSmoothing(train, seasonal_periods=12, trend='mul', seasonal='mul').fit()
```

```python
fit_1.params
```

```python
fit_1.fittedvalues
```

#### *Прогноз на 30 років*

```python
fit_1.forecast(491)
```

```python
ax = df.plot(figsize=(15,6), color='black', title="Прогноз методом Хольта Вінтерса" )
fit_1.fittedvalues.plot(ax=ax, style = '--', color='red')
fit_1.forecast(491).plot(ax=ax, style = '--', color='green')

```

### *Отримання метрик*

```python
metrics(train, fit_1.fittedvalues)
```

```python
metrics(test, fit_1.forecast(len(test)))
```

# Модель Sarimax

#### *Підбір параметрів для моделі*

```python
model_sr = auto_arima(train, seasonal=True, m=12, suppress_warnings=True,error_action='ignore', stepwise=True)
model_sr
```

#### *Cтворення і навчання моделі*

```python
sarimax_model = sm.tsa.statespace.SARIMAX(train, order=(0, 1, 1), seasonal_order=(1,0,1,12)).fit()
                                         
```

```python
sarimax_model.summary().tables[1]
```

#### *Діагностичні графіки моделі*

```python
sarimax_model.plot_diagnostics(figsize=(12, 10))
plt.show()
```

### *Прогноз*

#### *Прогноз та оцінка по тренувальним даним*

```python
predict=sarimax_model.get_prediction()
predict.predicted_mean
```

```python
predict=sarimax_model.get_prediction(start='1960-02-01')
metrics(train['1960-02-01':], predict.predicted_mean)
```

#### *Прогноз та оцінка по тестувальним даним*

```python
predict_test =sarimax_model.get_prediction(start='2008-01-01', end='2018-11-01')
predict_test.predicted_mean
```

```python
real_test = test['2008-01-01':'2018-11-01']['value']
metrics(real_test, predict_test.predicted_mean)
```

```python
ax = df.plot(figsize=(15,6), color='black', title="Прогноз методом SARIMA" )
sarimax_model.fittedvalues.plot(ax=ax, style='--', color='red')
predict.predicted_mean.plot(ax=ax, style='--', color='green')
plt.show()
```

*Висновок: Модель Хольта-Вінтерса у порівнянні з моделью Sarimax має більш низькі значення MAD и MSE, що в свою чергу свідчить про те, що модель Хольта-Вінтерса передбачає дані з меншою абсолютною та середньоквадратичною помилкою, що є позитивним показником точності моделі. Про точність моделі Хольта-Вінтерса свідчить також і менше значення стандартної похибки.   
    Разом з тим, показники MAPE і MPE для обох моделей невеликі, що свідчить про те, що моделі прогнозують дані з відносно невеликою похибкою.   
    Водночас слід зазначити, що обидві моделі показують досить низькі показники в тесті на стаціонарність. Це свідчить про те, що дані не є стаціонарними, що в свою чергу має негативний вплив на прогноз.*
