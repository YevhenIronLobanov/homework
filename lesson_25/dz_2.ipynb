### *_Опис рішення_*

1. На основі даних зазначених в умові задачі будуємо закон розподілу окремо для акцій X та акцій Y.
2. Розраховуємо окремо для акцій X та  акцій Y:  
    2.1 Очікувану середню норму прибутку (математичне очікування);  
    2.2 Дисперсію;

4. На основі отриманних даних розраховуємо:

    4.2 Коваріацію;  
    4.3 Кофіцієнт кореляції r_xy.
5. На основі показника коваріації та коефіцієнту кореляції робимо висновок щодо зв’язку між акціями X та Y.

### *_Рішення_*

### *Закон розподілу для величини X*
| X   |  10  |   5  |  2   |  1   |  -5  |  
|-----|------|------|------|------|------|    
| p   |  0.1 |  0.3 |  0.2 |  0.3 |  0.1 |

### *Закон розподілу для величини Y*

|  Y  |  20  |  10   |  2  |  -2  |  -8  |
|-----|------|------|------|------|------|
| p   | 0.1  |  0.3 |  0.2 |  0.3 |  0.1 |

```python
def math_exp (x, p):
    M = 0
    D = 0
    for i in range(len(x)):
        M += x[i]*p[i]
        D += ((x[i]**2)*p[i])
    D -= M**2
    σ = D**(1/2)
    return M, D, σ

M_X,D_X,σ_x = math_exp ([10, 5, 2, 1, -5], [0.1, 0.3, 0.2, 0.3, 0.1])
M_X = round(M_X, 3)
D_X = round(D_X, 3)
σ_x = round(σ_x, 3)
M_Y, D_Y, σ_Y = math_exp ([20, 10, 2, -2, -8], [0.1, 0.3, 0.2, 0.3, 0.1])
M_Y = round(M_Y, 3)
D_Y = round(D_Y, 3)
σ_Y = round (σ_Y, 3)

print('Очікувана середня норма прибутку по акціях X:', M_X,'%')
print('Очікувана середня норма прибутку по акціях Y:', M_Y,'%')
print ('D(X)=', D_X)                      
print('D(Y)=', D_Y)
print ('Ступінь ризику по акціях Х:', σ_x,'%')
print ('Ступінь ризику по акціях Y:', σ_Y,'%')
```

```python
import numpy as np
p = np.array([[0.01, 0.03, 0.02, 0.03, 0.01],
              [0.03, 0.09, 0.06, 0.09, 0.03],
              [0.02, 0.06, 0.04, 0.06, 0.02],
              [0.03, 0.09, 0.06, 0.09, 0.03],
              [0.01, 0.03, 0.02, 0.03, 0.01]])
x = np.array([10, 5, 2, 1, -5])
y = np.array([20, 10, 2, -2, -8])

```

```python
def math_all (x, y, p):
    M_XY = 0
    for i in range(len(x)):
        for j in range (len(y)):
            M_XY += x[i]*y[j]*p[i][j]
    return M_XY
M_XY = math_all(x, y, p)
print('Очікувана середня норма прибутку по акціях X та Y:', M_XY,'%')
```

```python
cov = M_XY-M_X*M_Y
cov = round(cov,3)
print('Коваріація:', cov)
r_xy = cov/(σ_x*σ_Y)
r_xy = round(r_xy,3)
print('Кореляція:', r_xy)
```
