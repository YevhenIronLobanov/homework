### *_Умова задачі_*

При вивченні випадкової величини Х у результаті 40 незалежних спостережень дістали вибірку:
10, 13, 10, 9, 9, 12, 12, 6, 7, 9, 8, 9, 11, 9, 14, 13, 9, 8, 8, 7, 10, 10, 11, 11, 11, 12, 8, 7, 9, 10, 14, 13, 8, 8, 9, 10, 11, 11, 12, 12. Побудувати дискретний статистичний розподіл для цієї вибірки, а також полігон частот і
F*(x). Знайти моду і медіану.

### *_Опис рішення_*

1. Створюємо функцію, яка приймає список елементів вибірки і за допомогою якої ми підраховуємо кількість раз коли зустрічається  унікальний елемент, що є в нашій вибірці. Функція створює словник в якому ключом є елемент вибірки, значенням - кількість раз коли зустрічається цей елемент.
2. На основі отриманих даних зі словника:  
    2.1 будуємо дискретний статистичний розподіл вибірки;  
    2.2 будуємо полігон частот.  
    2.3 розраховуємо емпіричну функцію розподілу F*(x) = n_x/n  
    2.4 будуємо емпіричну функцію розподілу.  
3. Розраховуємо моду і медіану.

### *_Рішення_*

```python
def distrib(lst):
    count_dict = {}
    for i in lst:
        if i in count_dict:
            count_dict[i] += 1
        else:
            count_dict[i] = 1
    return count_dict

            
```

```python
lst = [10, 13, 10, 9, 9, 12, 12, 6, 7, 9, 8, 9, 11, 9, 14, 13, 9, 8, 8, 7, 10, 10, 11, 11, 11, 12, 8, 7, 9, 10, 14, 13, 8, 8, 9, 10, 11, 11, 12, 12]

res = distrib(sorted (lst))
print(res)
```

```python
import matplotlib.pyplot as plt
numbers = list(res.keys())
frequencies = list(res.values())
plt.xlabel('Числа')
plt.ylabel('Частоти')
plt.title('Полігон частот')
plt.plot(numbers, frequencies, marker='o')
plt.show()
```

```python
import numpy as np
def empir_function (data, x):
    n = len(data)
    count = np.sum(data<=x)
    return count/n
sorted_data = np.sort(np.unique(lst))
x_val = sorted_data
empir_val = [empir_function(lst, x) for x in x_val]
for x, y in zip(x_val, empir_val):
    print(f'F*({x}) = {y}')
    
```

```python
plt.step(x_val, empir_val, where='post')
plt.xlabel('Числа')
plt.ylabel('F*(X)')
plt.title('Емпірична функція розподілу')

plt.show()
```

```python
def moda(count_dict):
    max_count = max(count_dict.values())
    for x, n_i in count_dict.items():
        if n_i == max_count:
            return x

result = moda(res)
print('Мо=',result)
```

```python
def mediana(lst):
    sorted_lst = sorted(lst)
    n = len(sorted_lst)
    if n % 2 == 0:
        mid = n // 2
        median = (sorted_lst[mid-1] + sorted_lst[mid]) / 2
    else:
        mid = n // 2
        median = sorted_lst[mid]
    return median
r = mediana(lst) 
print ('Ме=',r)
```
