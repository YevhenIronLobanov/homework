### *Визначаємо предиктори X для побудови модели*

*1. Предиктори X: 'Податок на бензин', 'Середній заробіток', 'Асфальтоване шосе', 'Водійські права'.  
   Шукома величина y: 'Споживання бензину'.  
2. Найвищий коефіцієнт кореляції між предикторами спостерігається між предиктором 'Податок на бензин' та 'Асфальтоване шосе' (-0.52). Разом з тим, предиктор 'Асфальтоване шосе' має дуже слабку кореляцію з 'Споживання бензину' (0.019), в свою чергу як предиктор 'Податок на бензин' має більший показник зв'язку з 'Споживання бензину' (зворотня кореляція = -0.45).
Враховуючи зазначене виключаємо предиктор ''Асфальтоване шосе'.*
   

```python
petrol_data = petrol[['Податок на бензин','Середній заробіток', 'Водійські права']].values
petrol_target = petrol[['Споживання бензину']].values
petrol_data_name = ['Податок на бензин', 'Середній заробіток', 'Водійські права']
X, y = scale(petrol_data), petrol_target
```

```python
missing = X==np.NAN
X[missing==True]
```

```python
model_petrol = LinearRegression()
model_petrol.fit(X, y)
```

```python
model_petrol.score(X, y)
```

```python
print(f'beta0:{model_petrol.intercept_}, beta:{model_petrol.coef_}')
```

```python
y_pred = model_petrol.predict(X)
print(f'Prediction {y_pred [1:7]}')
```

```python
plt.plot(X, y_pred, 'go')
plt.xlabel('Data(X)')
plt.ylabel('Target(y)')
plt.show()
```

```python
model_petrol.score(X, y_pred)
```

### *Оцінка якості лінійної регресії*

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error
from math import sqrt
```

```python
mae = mean_absolute_error(y, y_pred)
mse = mean_squared_error(y, y_pred)
rmse = sqrt(mse)
```

```python
print(f'Середня абсолютна похибка: {mae:.2f}')
print(f'Середня квадратична похибка: {mse:.2f}')
print(f'Корінь середньої квадратичної похибки: {rmse:.2f}')
```
