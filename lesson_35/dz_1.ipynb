```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

import sklearn
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import scale

from pylab import rcParams
rcParams['figure.figsize'] = 9,7
```

```python
df = 'student_scores.csv'
student = pd.read_csv(df)
student.head(10)
```

```python
plt.title ('Залежність між часом навчання і середнім балом студентів')
plt.xlabel('Час навчання')
plt.ylabel('Середній бал')
plt.plot(student['Hours'], student['Scores'], 'go')
```

```python
X = np.array([student['Hours']])
y = np.array ([student['Scores']])
X = X.reshape(-1,1)
y = y.reshape(-1,1)
```

```python
model = LinearRegression()
model.fit(X,y)
```

```python
print(f'alfa: {model.intercept_}, beta:{model.coef_}')
```

```python
print (f'Точність моделі: {model.score(X,y)}')
```

```python
X_new = np.array([10, 11, 12, 13, 14])
X_new = X_new.reshape(-1,1)
X_new
```

```python
y_new = model.predict(X_new)
y_new
```

```python
plt.title ('Залежність між часом навчання і середнім балом студентів')
plt.xlabel('Час навчання')
plt.ylabel('Середній бал')
plt.plot(student['Hours'], student['Scores'], 'go',label= 'Фактичний середній бал студентів'  )
plt.plot(X_new, y_new, 'ro', label= 'Прогнозний середній бал студентів' )
plt.legend()
```

```python
print (f'Точність лінійної регресії: {model.score(X_new,y_new)}')
```
